```python
import pandas as pd


ilmajaam_df = pd.read_csv("/content/Tallinn-Harku-2004-juuni-2024CSV.csv", sep=';')
ilmajaam_df.head(n=10)

#puhastan ebavajalikud read

ilmajaam_df = ilmajaam_df.drop(columns=['Õhurõhk merepinna kõrgusel hPa', 'Õhurõhk jaama kõrgusel hPa', 'Suhteline õhuniiskus %','Tunni miinimum õhutemperatuur °C', 'Tunni maksimum õhutemperatuur °C', '10 minuti keskmine tuule kiirus m/s' ])
ilmajaam_df
```

```python
import pandas as pd

# Convert 'Kuu' and 'Päev' to INT before concatenation
ilmajaam_df['Kuu'] = ilmajaam_df['Kuu'].astype(int)
ilmajaam_df['Päev'] = ilmajaam_df['Päev'].astype(int)

# Combine 'Aasta', 'Kuu', and 'Päev' columns to create a 'Kuupäev' column in 'YYYY-MM-DD' format
ilmajaam_df['Kuupäev'] = pd.to_datetime(ilmajaam_df['Aasta'].astype(str) + '-' +
                                         ilmajaam_df['Kuu'].astype(str) + '-' +
                                         ilmajaam_df['Päev'].astype(str),
                                         format='%Y-%m-%d', errors='coerce')

# Ensure 'Õhutemperatuur °C' is numeric
ilmajaam_df['Õhutemperatuur °C'] = pd.to_numeric(ilmajaam_df['Õhutemperatuur °C'], errors='coerce')

# Convert 'Tunni sademete summa mm' to numeric before calculating the mean
ilmajaam_df['Tunni sademete summa mm'] = pd.to_numeric(ilmajaam_df['Tunni sademete summa mm'], errors='coerce')

# Clean the 'Tunni maksimum tuule kiirus m/s' column:
ilmajaam_df['Tunni maksimum tuule kiirus m/s'] = (
    ilmajaam_df['Tunni maksimum tuule kiirus m/s']
    .astype(str)
    .str.replace(',', '.')
    .str.replace('[^0-9.]', '', regex=True)
)
ilmajaam_df['Tunni maksimum tuule kiirus m/s'] = pd.to_numeric(ilmajaam_df['Tunni maksimum tuule kiirus m/s'], errors='coerce')

# Group by the 'Kuupäev' and calculate the average for each parameter
daily_average = pd.DataFrame()
daily_average['Kuupäev'] = ilmajaam_df.groupby('Kuupäev')['Kuupäev'].first()  # Get unique dates
daily_average['Keskmine_temperatuur'] = ilmajaam_df.groupby('Kuupäev')['Õhutemperatuur °C'].mean()
daily_average['Keskmine_kiirgus'] = ilmajaam_df.groupby('Kuupäev')['Tunni keskmine summaarne kiirgus W/m²'].mean()
daily_average['Keskmine_sade'] = ilmajaam_df.groupby('Kuupäev')['Tunni sademete summa mm'].mean()
daily_average['Keskmine_suund'] = ilmajaam_df.groupby('Kuupäev')['10 minuti keskmine tuule suund °'].mean()
daily_average['Keskmine_max_tuule_kiirus'] = ilmajaam_df.groupby('Kuupäev')['Tunni maksimum tuule kiirus m/s'].mean()

# Reset the index to ensure it's a clean DataFrame
daily_average.reset_index(drop=True, inplace=True)

# Display the daily averages
print(daily_average)
```
